## 데이터 불러오기

```python
obesityDF = pd.read_csv("ObesityDataSet.csv")
obesityDF.head()
```

## encoding 하기

```python
obesityDF["CAEC"] = self_OrdinalEncoder(obesityDF, "CAEC", ["Always", "Frequently", "Sometimes", "no"])

obesityDF["CAEC"].value_counts()
```

```python
obesityDF["CALC"] = self_OrdinalEncoder(obesityDF, "CALC", ["Always", "Frequently", "Sometimes", "no"])

obesityDF["CALC"].value_counts()
```

```python
obesityDF["NObeyesdad"] = self_OrdinalEncoder(obesityDF, "NObeyesdad", [
        "Insufficient_Weight",
        "Normal_Weight",
        "Overweight_Level_I",
        "Overweight_Level_II",
        "Obesity_Type_I",
        "Obesity_Type_II",
        "Obesity_Type_III"])

obesityDF["NObeyesdad"].value_counts()
```

## 순서 상관 없는, 범주형 데이터만 encoding

```python

for idx, type_ in enumerate(obesityDF.dtypes):
    name = obesityDF.columns[idx]
    if type_ == "object" and obesityDF.columns[idx] != "MTRANS":
        obesityDF[name] = self_OrdinalEncoder(obesityDF, name, obesityDF[name].unique())
        print(obesityDF.iloc[:,idx].value_counts())
        print("\n\n")
# MTRANS는 마지막에 처리하기 
```

## float round() -> 키 몸무게 빼고

```python
for idx, type_ in enumerate(num_DF.dtypes):
    name = num_DF.columns[idx]
    if type_ == "float64" and name not in ["Height", "Weight"]:
        print(name)
        num_DF[name] = num_DF[name].round().astype(int)
        # print(obesityDF.iloc[:,idx].value_counts())
        print("\n")
```
